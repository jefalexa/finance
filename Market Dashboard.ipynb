```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import datetime as dt

def make_close_plot(df, name):
    sns.set_style('white')
    plt.figure(figsize= (18, 8))
    sns.lineplot(y=df['Close'], x=df.index)
    sns.despine(left=True)
    plt.axvline(dt.datetime.strptime('2020-03-11', '%Y-%m-%d'), ymax=0.95, linestyle='--')
    plt.text(dt.datetime.strptime('2020-03-11', '%Y-%m-%d'), df['Close'].max(), 'Pandemic Start', style='italic', fontsize=10)
    plt.xlabel('Date', fontsize=12, fontweight = 'semibold')
    plt.title(name, fontsize=18, fontweight = 'semibold')
    return(plt)

def show_ticker_symbols(tickers, period="1y"):
    for t in tickers:
        t_info = t.info
        t_name = f"{t_info['shortName']} ({t_info['symbol']})"
        t_df = t.history(period=period, rounding=True)
        plt = make_close_plot(t_df, t_name)
        plt.show()
    
```

```python
csco = yf.Ticker("CSCO")
soyb = yf.Ticker("SOYB")
corn = yf.Ticker("CORN")
gold = yf.Ticker("GLD")
oil = yf.Ticker("USO")

tickers = [csco, soyb, corn, gold, oil]
```

```python
show_ticker_symbols(tickers, '3y')
```
